```python
# check content trend over the years

df['release_year'].value_counts().sort_index().plot(kind='line',figsize=(10,5),title='Netflix content trend')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))  # Bigger figure size
ax = df['release_year'].value_counts().sort_index().plot(kind='bar', color='crimson')

plt.title("Netflix Content Trend Over the Years", fontsize=16)
plt.xlabel("Release Year", fontsize=14)
plt.ylabel("Number of Releases", fontsize=14)

#  Rotate x-axis labels & show every 5th year
plt.xticks(rotation=45)  
ax.set_xticks(range(0, len(df['release_year'].value_counts()), 5))  # Show every 5th year

plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.show()

```

```python
#Visualize the top genres using a bar graph:

plt.figure(figsize=(12, 6))
df['listed_in'].value_counts().head(10).plot(kind='barh', color='dodgerblue')

plt.title("Top 10 Most Common Netflix Genres", fontsize=16)
plt.xlabel("Number of Shows", fontsize=14)
plt.ylabel("Genre", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.0)

plt.show()
```

```python
#Find the Top 10 countries with the most Netflix shows/movies:

df['country'].value_counts().head(10)
```

```python
plt.figure(figsize=(12, 6))
df['country'].value_counts().head(10).plot(kind='bar', color='maroon')

plt.title("Top 10 Countries Producing Netflix Content", fontsize=16)
plt.xlabel("Country", fontsize=14)
plt.ylabel("Number of Shows/Movies", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.0)

plt.show()
```
